```python
import numpy as np
np.random.seed(5)
outliers=[]
outlier_free_list = []
def detect_outliers(data):
    
    threshold=2
    mean = np.mean(data)
    std =np.std(data)
    
    
    for i in data:
        z_score= (i - mean)/std 
        if np.abs(z_score) > threshold:
            outliers.append(i)
        else : 
            outlier_free_list.append(i)
    return outliers

outliers = detect_outliers(diabetes1.Target)
print("The outliers found in the Target variable :",outliers)
```

```python
import math
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

import the math library along with the mean_squared_error function from sklearn.metrics library 

```python
#Linear hypothesis function
def hypothesis(theta,X): 
    """function that multiplies the baseline attibutes that were chosen 
    with the initialized theta parameter 
    """
    return np.dot(X,theta)

#calculates the learning cost 
def computeCostMulti(X_train,Y_train,theta,m): #Cost function
    """
    theta is an n- dimensional vector of initial theta guess
    X is matrix with n- columns and m- rows
    y is a matrix with m- rows and 1 column
    """
    #note to self: *.shape is (rows, columns)
    J = float((1./(2*m)) * np.dot((hypothesis(theta,X_train)-Y_train).T,(hypothesis(theta,X_train)-Y_train)))
    return J

#Choose learning rate
alpha = 0.15
#Choose number of iterations
num_iters = 10000

# Initialize the theta
theta = np.zeros(8)

#Function for calculating the gardientDescent algorithm. 
def gradientDescentMulti(X_train, Y_train, theta, alpha, iterations,m):
    """
    Parameters
    X = ndarray, shape (n_samples, n_features)(m*n) = Training Data
    y = ndarray, shape (n_samples) = Labels
    theta = ndarray , shape(n_features) = n. Theta is an Initial linear regression parameter
    alpha = float, it's a learning rate
    
    """
    J_history = np.zeros(iterations)
    for i in range(num_iters):
        theta -= alpha / m * ((hypothesis(theta,X_train) - Y_train).T.dot(X_train))
        J_history[i] = computeCostMulti(X_train, Y_train, theta,m)

    return theta, J_history

#building the Linear Regression model 
def LR_model(X_train,Y_train,X_test,Y_test,theta,alpha,num_iters):
  """This is the function that defines the Linear Regression model,
  and trains and makes the prediction 
  """
  m = len(Y_train)
  X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
 
  print ('Calculate cost with Initial Theta value: ',computeCostMulti(X_train,Y_train,theta,m))
  print("-----------------------------------------------------------------------------------------------------------------------")
  theta, J_history = gradientDescentMulti(X_train, Y_train, theta, alpha, num_iters,m)
  # get the cost (error) of the model
  print ('Further testing and calculate theta after performing gradient decent:',computeCostMulti(X_train,Y_train,theta,m))
  print("-----------------------------------------------------------------------------------------------------------------------")
  print('theta:', theta.ravel())
  print("-----------------------------------------------------------------------------------------------------------------------")
  plt.plot(J_history)
  plt.xlabel('Iterations')
  plt.ylabel('Cost_J')
  plt.show()
  print("-----------------------------------------------------------------------------------------------------------------------")
  #make the predictions and store it in a list 
  pred_model_LR = []
  for i in X_test:
    i = np.hstack((np.ones(1), i))
    target = i.dot(theta)
    pred_model_LR.append(target)
  
  plt.figure(figsize=(10,10))
  plt.scatter(Y_test, pred_model_LR, c='crimson')
  plt.yscale('log')
  plt.xscale('log')

  p1 = max(max(pred_model_LR), max(Y_test))
  p2 = min(min(pred_model_LR), min(Y_test))
  plt.plot([p1, p2], [p1, p2], 'b-')
  plt.xlabel('Actual Values', fontsize=15)
  plt.ylabel('Predictions', fontsize=15)
  plt.axis('equal')
  plt.title('Predictions vs Actual LR')
  plt.show()
  print("-----------------------------------------------------------------------------------------------------------------------")
  #calculate the rmse 
  print("Rmse of the linear regression",math.sqrt(mean_squared_error(pred_model_LR,Y_test)))
  print("MAE of the linear regression",mean_absolute_error(pred_model_LR,Y_test))
```

```python
LR_model(X_train,Y_train,X_test,Y_test,theta,alpha,num_iters)
```

The above output shows the initial cost of the model without gradient descent which was 13521.7790 , and after performing gradient descent, the cost with the new theta is 1451.3938.
```python
def GBR_model(X_train,Y_train,X_test,Y_test):
  """This function defines the Gradient Boosting Regressor model and then makes a
  prediction based on the X_test dataset and the calculates the rmse 
  of the prediction made by the model
  """
  GBR_model = GradientBoostingRegressor(random_state=1, n_estimators=1000, max_depth=10).fit(X_train, Y_train)
  pred_model_GBR = GBR_model.predict(X_test)
  
  print("-----------------------------------------------------------------------------------------------------------------------")
  plt.figure(figsize=(10,10))
  plt.scatter(Y_test, pred_model_GBR, c='crimson')
  plt.yscale('log')
  plt.xscale('log')

  p1 = max(max(pred_model_GBR), max(Y_test))
  p2 = min(min(pred_model_GBR), min(Y_test))
  plt.plot([p1, p2], [p1, p2], 'b-')
  plt.xlabel('Actual Values', fontsize=15)
  plt.ylabel('Predictions', fontsize=15)
  plt.axis('equal')
  plt.title('Predictions vs Actual GBR')
  plt.show()
  print("-----------------------------------------------------------------------------------------------------------------------")
  print("Rmse of the Gradient Boosting Regressor",math.sqrt(mean_squared_error(pred_model_GBR,Y_test)))
  print("MAE of the linear regression",mean_absolute_error(pred_model_GBR,Y_test))
```
